```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/DOGE-USD.csv')
df
```

```python
df=df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python
df=df[['Close']]
df
```

```python
predict_day=1
df['Prediction']=df[['Close']].shift(-predict_day)
df
```

```python
df1=df.head(1000)
```

```python
x=np.array(df.drop(['Prediction'],1))
x=x[:len(df1)-predict_day -1]
x
```

```python
y=np.array(df1['Prediction'])
y=y[:-predict_day -1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor(n_estimators=2,random_state=585)
forest.fit(x_train,y_train)
print(forest.score(x_test,y_test))
```

```python
prediction=forest.predict(x_test)
print(prediction)
print()
print(y_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test)
plt.show()
```

```python
temp_df1=df1[:-predict_day]
x_val=temp_df1.tail(1)['Close'][0]
x_val
```

```python
predict_day=100
prediction=forest.predict([[x_val]])
print('predicted price of Dogecoin in ',predict_day,'days to be',prediction)
print('predicted actual price of Dogecoin in ',temp_df1.tail(1),['prediction'][0])
```

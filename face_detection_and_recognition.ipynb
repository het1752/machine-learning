```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import face_recognition
import numpy as np
i1=face_recognition.load_image_file('/content/dhoni1.jpg')
i1=cv2.cvtColor(i1,cv2.COLOR_BGR2RGB)
i2=face_recognition.load_image_file('/content/dhoni3.jpg')
i2=cv2.cvtColor(i2,cv2.COLOR_BGR2RGB)
facelocation=face_recognition.face_locations(i1)[0]
encodei1=face_recognition.face_encodings(i1)[0]
cv2.rectangle(i1,(facelocation[3],facelocation[0]),
(facelocation[1],facelocation[2]),(255,0,255),2)
facelocation=face_recognition.face_locations(i2)[0]
encodei2=face_recognition.face_encodings(i2)[0]
cv2.rectangle(i2,(facelocation[3],facelocation[0]),
(facelocation[1],facelocation[2]),(255,0,255),2)
result=face_recognition.compare_faces([encodei1],encodei2)
facedisc=face_recognition.face_distance([encodei1],encodei2)
print(result,facedisc)
cv2.putText(i1,f'{result} {round(facedisc[0],2)}',(50,50),
            cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
cv2_imshow(i1)
cv2_imshow(i2)

```
